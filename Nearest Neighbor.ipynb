```python
import numpy as np
import pandas as pd
```

```python
train_df = pd.read_csv('nnTrain.csv')
train_df
```

```python
test_df = pd.read_csv('nnTest.csv')
test_df
```

```python
n = train_df.shape[0]
m = test_df.shape[0]
print(n, " ", m)
dist = pd.DataFrame(np.zeros((n, m)))
for i in range(n):
    dist[0][i] = (train_df['var1'][i] - test_df['var1'][0]) ** 2 + \
               (train_df['var2'][i] - test_df['var2'][0]) ** 2 + \
               (train_df['var3'][i] - test_df['var3'][0]) ** 2

dist[0]
```

```python
n = train_df.shape[0]
m = test_df.shape[0]
print(n, " ", m)
dist = pd.DataFrame(np.zeros((n, m)))
for j in range(m):
    for i in range(n):
        dist[j][i] = (train_df['var1'][i] - test_df['var1'][j]) ** 2 + \
               (train_df['var2'][i] - test_df['var2'][j]) ** 2 + \
               (train_df['var3'][i] - test_df['var3'][j]) ** 2
dist.columns = ['test0', 'test1', 'test2', 'test3']
dist
```

```python
freq = train_df.iloc[dist.nsmallest(4, 'test3')['test3'].index]['class'].value_counts()
freq
```
